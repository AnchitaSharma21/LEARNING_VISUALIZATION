Question 9: Scatter Plot with Categories
Plot Height vs Weight categorized by gender.
Input Format:

160 55 M
165 60 F
170 68 M
158 52 F
172 70 M

```python
height = [160,165,170,158,172]
weight = [55,60,68,52,70]
gender = ["M","F","M","F","M"]
Mheight = []
Mweight = []
Fheight = []
Fweight = []
for i in range(len(gender)):
    if gender[i] == "M":
        Mheight.append(height[i])
        Mweight.append(weight[i])
    else:
        Fheight.append(height[i])
        Fweight.append(weight[i])

plt.scatter(Mheight, Mweight, label="Male",color="blue")
plt.scatter(Fheight, Fweight, label="Female",color="pink")

plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend()
plt.show()
```

Q11.Visualize the same dataset using multiple plots.

```python
data = [12, 15, 14, 16, 18, 21, 19, 17]
plt.figure(figsize=(15,5))

# Histogram
plt.subplot(1,3,1)
plt.hist(data)
plt.title("Histogram")

# Boxplot
plt.subplot(1,3,2)
sns.boxplot(data=data)
plt.title("Boxplot")

# Violin plot
plt.subplot(1,3,3)
sns.violinplot(data=data)
plt.title("Violin Plot")

plt.show()
```

Question 12: Folium Map Marker Create an interactive map and mark given location.

Input Format: 32.7266 74.8570 Jammu

```python
import folium

m = folium.Map(location=[32.7266, 74.8570], zoom_start=10)

folium.Marker([32.7266, 74.8570],popup="Jammu").add_to(m)
m.save("map.html")
m
```

Question 13: Multiple Location Mapping
Plot multiple city locations on a Folium map.
    
Input Format:
Jammu 32.7266 74.8570
Delhi 28.7041 77.1025
Chandigarh 30.7333 76.7794

```python
m = folium.Map(location=[30, 76], zoom_start=6)

cities = [("Jammu", 32.7266, 74.8570),
    ("Delhi", 28.7041, 77.1025),
    ("Chandigarh", 30.7333, 76.7794)]

for city, lat, lon in cities:
    folium.Marker([lat, lon],popup=city).add_to(m)

m
```
