```python
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
image_path = "/Users/USER/Downloads/image_image.jpg"
image = Image.open(image_path)
import matplotlib.pyplot as plt
plt.imshow(image) 
plt.axis('off') # Turn off axis numbers and ticks
plt.show()
```

CONVERTING TO GRAYSCALE

```python
grayscale_image = image.convert("L")   #Converts image to grayscale data
plt.imshow(grayscale_image, cmap='gray')  #Displays the data in gray colors
plt.axis('off')
plt.show()
```

```python
import numpy as np
```

```python
histogram = grayscale_image.histogram() #counts how many pixels exist at each intensity level.
pixel_values = np.arange(256)
plt.bar(pixel_values, histogram, color='blue', width=2.0)
plt.title('Grayscale Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 255])  # Ensure the x-axis covers the entire range
plt.show()
```

```python
#In digital images, “size” actually means the number of pixels.
resized_image = image.resize((image.width//20, image.height//20)) 
#What changed? Number of pixels/Resolution
plt.imshow(resized_image)
plt.axis('off')
plt.show()
```

```python
print(image.size)
print(resized_image.size)
```

```python
resized_image = image.resize((image.width*10, image.height*10)) 
plt.imshow(resized_image)
plt.axis('off')
plt.show()
```

```python
rotated_image = image.rotate(195)
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

```python
crop_box = (300, 350, 400, 400)
cropped_image = image.crop(crop_box)
plt.imshow(cropped_image)
plt.axis('off')
plt.show()
```

```python
from PIL import ImageEnhance
enhancer_contrast = ImageEnhance.Contrast(image) # Enhance contrast
contrast_image = enhancer_contrast.enhance(2)  # Enhance brightness
enhancer_brightness = ImageEnhance.Brightness(image)
brightness_image = enhancer_brightness.enhance(1.5)  #Enhance sharpness
enhancer_sharpness = ImageEnhance.Sharpness(image)
sharpness_image = enhancer_sharpness.enhance(20)  
plt.figure(figsize=(20, 5))
```

```python
# Original Image
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')

# High Contrast Image
plt.subplot(1, 4, 2)
plt.imshow(contrast_image)
plt.axis('off')
plt.title('High Contrast')

# High Brightness Image
plt.subplot(1, 4, 3)
plt.imshow(brightness_image)
plt.axis('off')
plt.title('High Brightness')

# High Sharpness Image
plt.subplot(1, 4, 4)
plt.imshow(sharpness_image)
plt.axis('off')
plt.title('High Sharpness')
plt.tight_layout()
plt.show()
```
